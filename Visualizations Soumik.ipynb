```python
fig1b=fig1[fig1.Subfigure=='B'].pivot(
    index='Element', columns='Attribute', values='Value')
```

```python
fig1b
```

```python
fig1b.index = fig1b.TaxonomyValue
fig1b = fig1b.drop(columns=['Color', 'TaxonomyValue', 'TaxonomyLevel'])
```

```python
fig1b = fig1b.apply(pd.to_numeric)
```

```python
fig1b.iloc[1].name
```

```python
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(7,6), sharey=True, sharex=True)
mycolors=['cyan','orange','green','red','purple','brown','gray','pink']
for i in range(4):
    for j in range(2):
        axs[i,j].barh(y=fig1b.iloc[2*i+j].index.str.replace(' Fraction', ''), 
        width=100*fig1b.iloc[2*i+j].values, color=mycolors[2*i+j])
        axs[i,j].set_title(fig1b.iloc[2*i+j].name, size=11)
fig.tight_layout()
fig.text(x=0.5, y=0, s='Percentage of Non-US PhDs')
fig.show()
```

```python
fig1c=fig1[fig1.Subfigure=='C'].pivot(index='Element', columns='Attribute', values='Value')
```

```python
fig1c.TaxonomyLevel.value_counts()
```

fig1c = fig1c.drop(columns=['P', 'TaxonomyLevel'])

```python
fig1c_taxonomy = pd.read_csv('us-faculty-hiring-and-retention-code/data/taxonomization/clean-ftas-taxonomy-hierarchy.csv')
```

```python
fig1c_taxonomy = fig1c_taxonomy.loc[fig1c_taxonomy.Umbrella.isin(["Applied Sciences", "Education", "Engineering", "Humanities", "Mathematics and Computing", "Medicine and Health", "Natural Sciences", "Social Sciences"])]
```

```python
fig1c['Field'] = fig1c.TaxonomyValue.map(fig1c_taxonomy.set_index('Taxonomy').Umbrella.to_dict())
```

```python
fig1c = fig1c.dropna()
```

```python
TaxonomyValue_mapping = {"Applied Sciences": 1, "Education": 2, "Engineering": 3, "Humanities": 4, "Mathematics and Computing": 5, "Medicine and Health" : 6, "Natural Sciences" : 7, "Social Sciences" : 8}
fig1c.loc[:,'Field'] = fig1c['Field'].map(TaxonomyValue_mapping)
```

```python
fig1c
```

```python
fig1c['P'] = fig1c['P'].apply(pd.to_numeric)
```

```python
mycolors=['cyan','orange','green','red','purple','brown','gray','pink']
for i in range(1,9):
    plt.scatter(fig1c[fig1c.Field==i].Field, (fig1c[fig1c.Field==i]['P']), marker='x', color=mycolors[i-1])
```
